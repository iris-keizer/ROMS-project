## Notebook to prepare the ROMS output data from Dewi to be able to save the datasets locally

The next step would be to create the datasets for lateral forcing.

In this notebook:
- Datetime objects are changed into years


The ROMS output data from Dewi is:
- 1993-2013
- monthly
- 1/8°x1/8°,

The data is downloaded from: 



 

```python
# Import necessary packages

import ast
import numpy as np
import xarray as xr

from tqdm import tqdm
```

```python
# Import ROMS output dewi
roms_output_dewi4_exp29 = xr.open_dataset('/Users/iriskeizer/Documents/ROMS/data/ROMS output Dewi/NorthSea4_exp29_zeta.nc')
roms_output_dewi4_exp36 = xr.open_dataset('/Users/iriskeizer/Documents/ROMS/data/ROMS output Dewi/NorthSea4_exp36_zeta.nc')
roms_output_dewi8 = xr.open_dataset('/Users/iriskeizer/Documents/ROMS/data/ROMS output Dewi/NorthSea8_exp03_zeta.nc')

# Obtain annual averages
roms_output_dewi4_exp29 = roms_output_dewi4_exp29.groupby('ocean_time.year').mean()
roms_output_dewi4_exp36 = roms_output_dewi4_exp36.groupby('ocean_time.year').mean()
roms_output_dewi8 = roms_output_dewi8.groupby('ocean_time.year').mean()
```

```python
roms_output_dewi4_exp29.zeta.plot(col = 'year', col_wrap = 7)
```

```python
roms_output_dewi4_exp36.zeta.plot(col = 'year', col_wrap = 7)
```

```python
roms_output_dewi8.zeta.plot(col = 'year', col_wrap = 7)
```

## Obtain data sets with lat and lon as dimensions

```python
lats_4 = [lst[0] for lst in roms_output_dewi4_exp29.lat_rho.values]
lons_4 = list(roms_output_dewi4_exp29.lon_rho.values[0])

lats_8 = [lst[0] for lst in roms_output_dewi8.lat_rho.values]
lons_8 = list(roms_output_dewi8.lon_rho.values[0])
```

```python
ROMS_output_dewi4_exp29 = xr.Dataset(
    data_vars = dict(
        slh = (["year", "lat", "lon"], roms_output_dewi4_exp29.zeta.values)
    ),
    coords = dict(
        year = (["year"], roms_output_dewi4_exp29.year.values),
        lat = (["lat"], lats_4),
        lon = (["lon"], lons_4),
    ),
    attrs = dict(
        description = "ROMS slh output from Dewi with 1/4° resolution exp29",
    )    
)
    
ROMS_output_dewi4_exp29.slh.attrs['units'] = "m"
ROMS_output_dewi4_exp29.slh.attrs['long name'] = "Sea level height above the geoid"
```

```python
ROMS_output_dewi4_exp36 = xr.Dataset(
    data_vars = dict(
        slh = (["year", "lat", "lon"], roms_output_dewi4_exp36.zeta.values)
    ),
    coords = dict(
        year = (["year"], roms_output_dewi4_exp36.year.values),
        lat = (["lat"], lats_4),
        lon = (["lon"], lons_4),
    ),
    attrs = dict(
        description = "ROMS slh output from Dewi with 1/4° resolution exp36",
    )    
)
    
ROMS_output_dewi4_exp36.slh.attrs['units'] = "m"
ROMS_output_dewi4_exp36.slh.attrs['long name'] = "Sea level height above the geoid"
```

```python
ROMS_output_dewi8 = xr.Dataset(
    data_vars = dict(
        slh = (["year", "lat", "lon"], roms_output_dewi8.zeta.values)
    ),
    coords = dict(
        year = (["year"], roms_output_dewi8.year.values),
        lat = (["lat"], lats_8),
        lon = (["lon"], lons_8),
    ),
    attrs = dict(
        description = "ROMS slh output from Dewi with 1/8° resolution",
    )    
)
    
ROMS_output_dewi8.slh.attrs['units'] = "m"
ROMS_output_dewi8.slh.attrs['long name'] = "Sea level height above the geoid"
```

```python
ROMS_output_dewi4_exp29.slh.plot(col = 'year', col_wrap = 7)
```

```python
ROMS_output_dewi4_exp36.slh.plot(col = 'year', col_wrap = 7)
```

```python
ROMS_output_dewi8.slh.plot(col = 'year', col_wrap = 7)
```

### Save the datasets

```python
ROMS_output_dewi4_exp29.to_netcdf(f"/Users/iriskeizer/Documents/ROMS/data/ROMS output Dewi/ROMS_output_dewi4_exp29.nc")
ROMS_output_dewi4_exp36.to_netcdf(f"/Users/iriskeizer/Documents/ROMS/data/ROMS output Dewi/ROMS_output_dewi4_exp36.nc")
ROMS_output_dewi8.to_netcdf(f"/Users/iriskeizer/Documents/ROMS/data/ROMS output Dewi/ROMS_output_dewi8.nc")
```
