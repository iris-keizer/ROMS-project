## Notebook to prepare the GLORYS12V1 data to be able to save the datasets locally

The next step would be to create the datasets for lateral forcing.

In this notebook:
- The seperate files for each variable are combined into 1 file


The GLORYS12V1 data is:
- 1993 - 2019
- monthly
- 1/12° x 1/12°

The GLORYS12V1 data is downloaded from https://resources.marine.copernicus.eu/product-detail/GLOBAL_MULTIYEAR_PHY_001_030/INFORMATION


 

```python
# Import necessary packages

import xarray as xr
```

```python
# Import GLORYS12V1 reanalysis data
glorys12v1_monthly = xr.open_dataset('/Users/iriskeizer/Documents/ROMS/data/Glorys12v1/SSH/cmems_mod_glo_phy_my_0.083_P1M-m_1648030292571.nc')


# Obtain annual averages
glorys12v1 = glorys12v1_monthly.groupby('time.year').mean('time')


# Drop 2019
glorys12v1 = glorys12v1.where(glorys12v1.year < 2019, drop=True)


# Save annual data
glorys12v1.to_netcdf(f'/Users/iriskeizer/Documents/ROMS/data/GLORYS12V1/SSH/glorys12v1_annual.nc')
```
