# Check global weigted average of ORA and AVISO

```python
import xarray as xr
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Import AVISO global altimetry data

aviso = pd.read_csv('/Users/iriskeizer/Documents/ROMS/data/AVISO/MSL_Serie_MERGED_Global_AVISO_GIA_NoAdjust_Filter2m.txt', header = None, sep = " ", names = ['Time', 'MSL', 'Nans'], index_col = 'Time').drop('Nans', axis=1)
```

```python
aviso = aviso*100 # Convert to cm
```

```python

aviso.plot()

plt.xlabel('Time [yr]', fontsize = 14)
plt.ylabel('Sea level [cm]', fontsize = 14)
```

```python
aviso.to_xarray()
```

```python
# Import ORA data
path = '/Users/iriskeizer/Documents/ROMS/data/ORA20C/opa2/zos/'

ora = xr.open_mfdataset(f'{path}zos_*')

ora = ora.drop('time_counter_bnds')

ora = ora.where(ora.time.dt.year > 1949, drop = True)
```

```python
# Obtain weighted global average
weights = np.cos(np.deg2rad(ora.lat))

ora_weighted = ora.weighted(weights)

ora_global = ora_weighted.mean(['lat', 'lon'])*100

ora_global_year = ora_global.groupby('time.year').mean('time')
```

```python
plt.figure()

ora_global.zos.plot()

plt.title('Monthly sea level of weighted global average', fontsize = 14)
plt.xlabel('Time [yr]', fontsize = 14)
plt.ylabel('Sea level [cm]', fontsize = 14)

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.savefig('/Users/iriskeizer/Documents/ROMS/figures/analysis/global/ora_monthly')
```

```python
plt.figure()

ora_global_year.zos.plot()

plt.title('Annual sea level of weighted global average', fontsize = 14)
plt.xlabel('Time [yr]', fontsize = 14)
plt.ylabel('Sea level [cm]', fontsize = 14)

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.savefig('/Users/iriskeizer/Documents/ROMS/figures/analysis/global/ora_annual')
```
