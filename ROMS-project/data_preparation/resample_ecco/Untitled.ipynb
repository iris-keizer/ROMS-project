# Compare ECCO data

I've both downloaded resampled ECCO data and resampled the ECCO data myself. Here, we compare those results.

```python
import xarray as xr
```

```python
path = '/projects/0/einf2878/ROMS/data/ECCO'
ssh_nctiles = xr.open_mfdataset(f'{path}/nctiles/sshdyn/*.nc')
ssh_resampled = xr.open_dataset(f'{path}/resampled/sshdyn.nc')
ssh_interp = xr.open_mfdataset(f'{path}/interp_monthly/SSHDYN/*.nc')
```
