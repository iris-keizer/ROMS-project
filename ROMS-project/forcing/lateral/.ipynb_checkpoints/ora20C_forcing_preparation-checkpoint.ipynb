## Change ORA20C to same format as GLORYS

First using ensemble member opa0

```python
import xarray as xr
```

```python
ora = xr.open_dataset('/Users/iriskeizer/Documents/ROMS/data/ORA20C/opa1_prepared.nc')
glorys = xr.open_dataset('/Users/iriskeizer/Documents/ROMS/data/GLORYS12V1/processed/glorys.nc')
```

```python
# Change dimension name of depth
ora = ora.rename({'dep':'depth', 'lon':'longitude', 'lat':'latitude'})
```

```python
ora.thetao.sel(time = ora.time.values[10]).sortby(['depth', 'longitude', 'latitude']).plot(col = 'depth', col_wrap = 7)
```

```python
# Assign same attributes to ORA-20C data

ora.so.attrs = {'long_name': 'Salinity', 'standard_name':'sea_water_salinity', 'units':'1e-3', 'unit_long':'Practical Salinity Unit', 'cell_methods':'area: mean'}
ora.thetao.attrs = {'long_name': 'Temperature', 'standard_name':'sea_water_potential_temperature', 'units':'degrees_C', 'unit_long':'Degrees Celsius', 'cell_methods':'area: mean',
                   '_CoordinateAxes': 'time depth lat lon',  'associate': 'time depth lat lon', 'interval_write': 'monthly', 'offline_operation': 'time average and spatial interpolation'}
ora.uo.attrs = {'long_name': 'Eastward velocity', 'standard_name':'eastward_sea_water_velocity', 'units':'m s-1', 'unit_long':'Meters per second', 'cell_methods':'area: mean'}
ora.vo.attrs = {'long_name': 'Northward velocity', 'standard_name':'northward_sea_water_velocity', 'units':'m s-1', 'unit_long':'Meters per second', 'cell_methods':'area: mean'}
ora.zos.attrs = {'long_name': 'Sea surface height', 'standard_name':'sea_surface_height_above_geoid', 'units':'m', 'unit_long':'Meters', 'cell_methods':'area: mean'}
ora.time.attrs = {'axis': 'T', 'long_name':'Time (hours since 1950-01-01)', 'standard_name':'time', '_CoordinateAxisType':'Time'}
ora.longitude.attrs = {'units':'degrees_east','unit_long':'Degrees East', 'long_name':'Longitude', 'standard_name':'longitude', 'axis':'X'}
ora.latitude.attrs = {'units':'degrees_north','unit_long':'Degrees North', 'long_name':'Latitude', 'standard_name':'latitude', 'axis':'Y'}

```

```python
ora.to_netcdf('/Users/iriskeizer/Documents/ROMS/code/model2roms_ora20c/input/ora20c.nc', format = "NETCDF4")
```

```python
ora.thetao.sel(time = ora.time.values[0]).sortby(['depth', 'longitude', 'latitude']).plot(col = 'depth', col_wrap = 7)
```
