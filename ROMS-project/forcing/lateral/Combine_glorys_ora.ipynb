## Notebook to get started on combining the GLORYS and ORA-20C data

Variables: 
We nudged temperature and salinity toward
outgoing nudging at a time scale of 1 year and ingoing nudging at a time scale of 1 day, using a 20‐point
relaxation zone at the lateral boundaries of the domain. Baroclinic velocities were nudged at the boundaries
only. To suppress computational noise, we also applied a 20‐point sponge layer, multiplying the diffusion
and viscosity by a factor that linearly increases to 10 at the lateral boundaries.


The GLORYS data covers 1993-2018 and has a resolution of 1/12°


The ORA-20C data covers 1900-2009 and has a spatial resolution of 1° and monthly time resolution
The ORA-20C data consists of 10 ensemble members, currently, only opa1 (a single member) is used
