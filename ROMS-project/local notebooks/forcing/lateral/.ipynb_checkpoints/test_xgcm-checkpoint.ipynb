## Linear vertical interpolation using xgcm

https://xgcm.readthedocs.io/en/latest/api.html?highlight=transform#xgcm.Grid.transform

```python
data
```

### depth to level

```python
xgcm_grid = Grid(data, coords={'Z': {'center': 'depth'},
                        },
                periodic=False
            )
xgcm_grid
```

```python
target_depth_levels = np.arange(0,500,20)
```

```python
temp_on_level = xgcm_grid.transform(data.thetao, 'Z', target_depth_levels, target_data=None, method='linear')
```

```python
temp_on_level.isel(time=0).plot(col='depth', col_wrap = 10)
```

## I've been performing a transformation with xgcm but I should try to perform an interpolation

```python
# Create xgcm grid object
xgcmgrid = Grid(input_data, coords={'X': {'center': 'xi_rho', 'center': 'xi_v', 'inner': 'xi_u'},
                                'Y': {'center': 'eta_rho', 'center': 'eta_u', 'inner': 'eta_v'},
                                'Z': {'center': 's_rho', 'center': 'depth', 'outer': 's_w'}},
                 periodic=False)
```

```python
output_data
```

```python
# Interpolate the vertical levels
output_data = data.copy()
output_data['depth'] = -1*output_data['depth']

output_data = output_data.drop(['z_u', 'z_v', 'z_w', 's_w', 'xi_v', 'eta_v', 'xi_u', 'eta_u', 'lat', 'lon'])

# Transpose input_data
#output_data['thetao'] = output_data['thetao'].transpose('time', 'depth', 'eta_rho', 'xi_rho')#.fillna(hmin/150)


# Create xgcm grid object
xgcmgrid = Grid(output_data, coords={'X': {'center': 'xi_rho'},
                                'Y': {'center': 'eta_rho'},
                                'Z': {'center': 'depth', 'inner': 's_rho'}},
                 periodic=False)

output_data['thetao'] = xgcmgrid.interp(output_data['thetao'], 'Z')

output_data.thetao.isel(time=0).plot(col='s_rho', col_wrap = 10)
```
