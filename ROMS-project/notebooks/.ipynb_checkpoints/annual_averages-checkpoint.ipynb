```python
path = '/projects/0/einf2878/output'
directory = 'SODA/spinup/run4'
file_name = 'NorthSea4_avg.nc'
```

```python
data_2019 = xr.open_dataset(f'{path}/{directory}/{file_name}')
```

## Some data processing

Make sure the attributes are kept

```python
# Sort coordinates in increasing order
data_2019 = data_2019.sortby(['ocean_time'])
data_2020 = data_2020.sortby(['ocean_time'])
```

```python
# Remove incomplete years
#data = data.where(data.ocean_time.dt.year < data.ocean_time.dt.year.data[-1], drop = True)
```

```python
#data.ocean_time
```

## Obtain annual averages

Make sure the attributes are kept

```python
data_annual_2019 = data_2019.groupby(data_2019.ocean_time.dt.year).mean('ocean_time', keep_attrs = True)
data_annual_2020 = data_2020.groupby(data_2020.ocean_time.dt.year).mean('ocean_time', keep_attrs = True)
```

```python
data_annual_2019 = data_annual_2019.rename({'year':'time'})
data_annual_2020 = data_annual_2020.rename({'year':'time'})
```

```python
data_annual = xr.concat([data_annual_2019, data_annual_2020], dim = 'time')
```

```python
data_annual
```

## Save dataset

```python
directory = 'SODA/spinup/run4/merge'
data_annual.to_netcdf(f'{path}/{directory}/NorthSea4_avg_run4_annual.nc')
```
