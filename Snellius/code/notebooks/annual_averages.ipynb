```python
import xarray as xr
```

## Import data

```python
path = '/projects/0/einf2878/output'
directory = 'GLORYS/spinup2'
file_name = 'NorthSea4_avg.nc'
```

```python
data = xr.open_dataset(f'{path}/{directory}/{file_name}')
```

## Some data processing

Make sure the attributes are kept

```python
# Sort coordinates in increasing order
data = data.sortby(['ocean_time'])
```

```python
# Remove incomplete years
#data = data.where(data.ocean_time.dt.year < data.ocean_time.dt.year.data[-1], drop = True)
```

```python
data.ocean_time
```

## Obtain annual averages

Make sure the attributes are kept

```python
data_annual = data.groupby(data.ocean_time.dt.year).mean('ocean_time', keep_attrs = True)
```

```python
data_annual = data_annual.rename({'year':'time'})
```

```python
data_annual
```

## Save dataset

```python
data_annual.to_netcdf(f'{path}/{directory}/NorthSea4_avg_spinup2_annual.nc')
```
